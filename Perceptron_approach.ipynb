```python
#Actual Work Starts Here
#Defined the pediction function
def perception_training(x):
    #took both the initial weights as 0
    w1 = 0
    w2 = 0
    #initial bias as 1
    t = 1
    #learning rate
    n = 1.5
    #y_hat initialized to all zeros
    y_hat = np.zeros(20)
    count = 0
    w_list = []

    #The while loop runs till either we find all the errors to be zero or we run 50 epochs
    while(True):
        error = 0
        #the for loops goes through all the 20 input x1, x2 and y
        for i in range(0, 20):
            #calculating the y_hat
            y_hat[i] = 1 if ((w1*x[i][0]) + (w2*x[i][1]) + t) >= 0 else 0
            #calculating del_w1, del_w2 and del_t and then updating the w1,w2 and t
            del_w1 = n*(x[i][2] - y_hat[i])*x[i][0]
            del_w2 = n*(x[i][2] - y_hat[i])*x[i][1]
            del_t = n*(x[i][2] - y_hat[i])
            w1 = w1 + del_w1
            w2 = w2 + del_w2
            t = t + del_t
            #calculating error in each step and updating the total error
            error = error + ((x[i][2] - y_hat[i])**2)
            w_list.append([w1, w2, t])
        #if total error is 0, we end the while loop
        if (error == 0):
            #print(error)
            break
        count = count + 1
        #if we ran a total of 50 epochs we end the while loop
        if (count > 50):
            break
    #print(w1, w2, t)
    #return the value of w1, w2, t
    return w1, w2, t, w_list
        
        
```

```python
df2_arr = df2.as_matrix()
```

```python
w1, w2, t, w_list = perception_training(df2_arr)
```
